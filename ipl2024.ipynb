```python
#Get highest_Bid (top 10) players with their team names and create a bar chart.
```

```python
Highest_Bid = ipl[['PLAYER_NAME','PRICE']].sort_values(by = 'PRICE', ascending=False).head(10)
```

```python
Highest_Bid = Highest_Bid.reset_index(drop=True)
Highest_Bid
```

```python
df = pd.DataFrame(ipl)
```

```python
plt.figure(figsize=(15,4))
ax = sns.barplot(x='PLAYER_NAME', y='PRICE', data = Highest_Bid , color = 'c')
for p in ax.patches:
    ax.annotate(f'{p.get_height():,.0f}',  # Label the height of the bar (price)
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Position of label
                ha='center', va='bottom',  # Align the label at the top of the bar
                fontsize=8, color='black')
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('Name of Players')
plt.ylabel('Price')
plt.title('Highest_Bid')
```

```python
Team_Runs = pd.concat([Batsman, Allrounder])
Team_Runs = Team_Runs.groupby('TEAM_NAME')['RUNS'].sum()
Team_Runs = Team_Runs.reset_index()
Team_Runs
```

```python
plt.figure(figsize=(23, 6))

# Create a line plot instead of a bar plot
dx = sns.lineplot(x='TEAM_NAME', y='RUNS', data=Team_Runs, marker='o', color='m', linewidth=2)

# Annotate the points with the values
for x, y in zip(Team_Runs['TEAM_NAME'], Team_Runs['RUNS']):
    dx.text(x, y + 5, f'{y}', ha='center', fontsize=14, color='black')

# Customize the axes and title
plt.xlabel('Name of Teams')
plt.ylabel('Runs')
plt.title('Team Runs')
```
